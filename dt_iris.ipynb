```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree 
from sklearn import metrics

from matplotlib import pyplot as plt
```

```python
iris = load_iris()
df1 = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df1['target'] = iris.target
```

```python
df1
```

```python
iris.target_names
```

```python
# Replacing class names for better visualization
df1['target'].replace(0, iris.target_names[0], inplace=True)
df1['target'].replace(1, iris.target_names[1], inplace=True)
df1['target'].replace(2, iris.target_names[2], inplace=True)

df1
```

```python
# split dataset in features and outcome
X = df1[iris.feature_names]
y = df1.target  # target
```

```python
# Split dataset into Training and Test (70% Training    30% Test)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
# Creating a Decision Tree
dt = DecisionTreeClassifier(criterion='entropy')

# Train the classifier
dt.fit(X_train, y_train)
```

```python
# predict values
y_pred = dt.predict(X_test)
```

```python
# Model Accuracy
print(f'Accuracy: {metrics.accuracy_score(y_test, y_pred)}')
```

```python
fig = plt.subplots(figsize=(12, 10))
plot_tree(dt,
          feature_names= iris.feature_names,
          class_names= iris.target_names,
          impurity=True,
          filled=True,
          rounded=True)
plt.show()
```
